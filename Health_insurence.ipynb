```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Health_Insurence.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# drop the  missing columns
df.dropna(subset=['age','bmi'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
# Encoding categorical variables
categorical_cols = ['sex', 'hereditary_diseases', 'city', 'job_title']
numerical_cols = ['age', 'weight', 'bmi', 'no_of_dependents', 'smoker', 'bloodpressure', 'diabetes', 'regular_ex']
```

```python
categorical_cols
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer([ ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)])
```

```python
preprocessor
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['claim'], bins=30, kde=True)
plt.title("Distribution of Insurance Claims")
plt.xlabel("Claim Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['smoker'], y=df['claim'])
plt.title("Impact of Smoking on Claim Amount")
plt.xlabel("Smoker (0=No, 1=Yes)")
plt.ylabel("Claim Amount")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['bmi'], y=df['claim'], hue=df['diabetes'])
plt.title("BMI vs Claim Amount (Colored by Diabetes Status)")
plt.xlabel("BMI")
plt.ylabel("Claim Amount")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['sex'], y=df['claim'])
plt.title("Impact of Gender on Claim Amount")
plt.xlabel("Gender")
plt.ylabel("Claim Amount")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Apply Label Encoding
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Now compute correlation
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x=df['diabetes'], y=df['claim'], hue=df['sex'], split=True)
plt.title("Claim Amount Distribution by Diabetes and Gender")
plt.xlabel("Diabetes (0=No, 1=Yes)")
plt.ylabel("Claim Amount")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['claim'])
y=df['claim']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Random Forest Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
```

```python
model=Pipeline([
    ('preprocessor',preprocessor),
    ('regrssor',RandomForestRegressor(n_estimators=100,random_state=42))])
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',LinearRegression())
])
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',DecisionTreeRegressor(criterion='squared_error',random_state=42))
])
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
sample_input = x_test.iloc[:5]  # Selecting first 5 test samples
y_sample_pred = model.predict(sample_input)

print("\nDemo Test Predictions:")
for i, pred in enumerate(y_sample_pred):
    print(f"Sample {i+1}: Predicted Claim Amount = {pred}")
```
